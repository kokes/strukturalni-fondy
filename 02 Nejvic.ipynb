# 2. Největší příjemci
Data jsou zpracována, je čas podívat se na nějaké statistiky.

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

## Načítání dat
Data od MMR máme zkombinována s daty of MFČR v souboru `vystupy/projekty-bez-popisu.csv` (můžeme použít i `projekty.csv`, ale ty jsou zbytečně velké pro naše potřeby).

```python
df = pd.read_csv('vystupy/projekty-bez-popisu.csv', low_memory=False)
```

```python
df.shape
```

```python
# nazvy sloupcu, at se to lip filtruje
prog = 'Program'
ico = 'IČO'
dat = 'Datum'
naz = 'Název'
celk = 'Celkové zdroje'
sml = 'Smlouva'
prop = 'Proplaceno'
zad = 'Žadatel'
prfo = 'Právní forma'
prfok = 'Právní forma (kód)'
nuts = 'NUTS4'
esa = 'ESA'
esak = 'ESA (kód)'
zam = 'Zaměstnanců'
kat = 'Kategorie'
ver = 'Veřejná instituce'
```

```python
df.loc[:, ico] = [str(j).rjust(8, '0') for j in df.loc[:, ico]]
```

## Absolutně největší příjemci

```python
absn = df.groupby([ico, zad]).sum().sort_values(by=sml, ascending=False).loc[:, [celk, sml, prop]]
(absn/10**9).round(2).head(15)
```

```python
(absn/10**9).round(2).head(200).to_csv('vystupy/tabulky/nejvetsi-prijemci.csv')
```

## Největší veřejní příjemci
Stejná tabulka, protože první neveřejný subjekt jsou 37. Třinecké železárny.

```python
absv = df.loc[df.loc[:, ver] == True].groupby([ico, zad]).sum().sort_values(by=sml, ascending=False).loc[:, [celk, sml, prop]]
(absv/10**9).round(2).head(15)
```

```python
(absv/10**9).round(2).head(200).to_csv('vystupy/tabulky/nejvetsi-prijemci-verejni.csv')
```

## Největší soukromí příjemci

```python
abss = df.loc[df.loc[:, ver] == False].groupby([ico, zad]).sum().sort_values(by=sml, ascending=False).loc[:, [celk, sml, prop]]
(abss/10**9).round(2).head(15)
```

```python
(abss/10**9).round(2).head(200).to_csv('vystupy/tabulky/nejvetsi-prijemci-soukromi.csv')
```

## Největší dotace
Největší individuální dotace.

```python
nejd = df.groupby([ico, zad, naz]).sum().loc[:, [celk, sml, prop]].sort_values(by=sml, ascending=False)
(nejd/10**9).head(15).round(2)
```

```python
(nejd/10**9).head(200).round(2).to_csv('vystupy/tabulky/nejvetsi-dotace.csv')
```
